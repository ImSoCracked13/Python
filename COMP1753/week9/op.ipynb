```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
n = int(input('Enter a number: '))
print(f'{n}! = {factorial(n)}')
```

```python
def rec_sum():
    if n == 1:
        return 1
    else:
        n = n + rec_sum(n - 1)
        return n
```

```python
def pyramid(n):
    if n == 1:
        return 1
    return pyramid(n-1) + n*n
pyramid()
```

```python
def fibonacci(n):
    if n == 0 or n == 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
n = 40
print(f'{n}: {fibonacci(n)}')
```

```python
import os
def list_files(path):
    if os.path.isdir(path):
        print(f'+ {path}')
        files = os.listdir(path)
        for f in files:
            list_files(path + '/' + f)
    else:
        print(f'- {path}')
```

```python
path = input('Enter a path: ')
list_files(path)
```
